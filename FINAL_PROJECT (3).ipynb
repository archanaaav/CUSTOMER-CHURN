```python
Q1=np.percentile(df['CreditScore'],25)
Q3=np.percentile(df['CreditScore'],75)


IQR= Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower,upper)
```

```python
outlier=[]
for x in df['CreditScore']:
    if(x>upper)or(x<lower):
        outlier.append(x)
outlier
```
